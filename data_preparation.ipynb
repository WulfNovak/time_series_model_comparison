- Due to day seasonality, creating features at the day level

```python
# 10 min interval and hour level daily averages to better capture daily fluctations

vars_to_avg = ['T1', 'RH_1', 'T2', 'RH_2', 'T3', 'RH_3', 'T4', 'RH_4', 
               'T5', 'RH_5', 'T6', 'RH_6', 'T7', 'RH_7', 'T8', 'RH_8', 
               'T9', 'RH_9', 'T_out', 'Press_mm_hg', 'RH_out', 'Windspeed', 
               'Visibility', 'Tdewpoint']
    
# Derive 'hour' and 'hour_minute' columns
energy_dt_2['hour'] = energy_dt_2.index.hour # dt.strftime('%H')
energy_dt_2['hour_minute'] = energy_dt_2.index.strftime('%H:%M')
    
# Group by 'hour' and calculate the average of the provided variables
hour_avg = energy_dt_2.groupby(['hour'])[vars_to_avg].mean()#.reset_index(level=1)
hour_avg.columns = ['h_avg_' + x for x in list(hour_avg.columns)]

# # Group by 'hour_minute' and calculate the average of the provided variables
hour_minute_avg = energy_dt_2.groupby(['hour_minute'])[vars_to_avg].mean()
hour_minute_avg.columns = ['hm_avg_' + x for x in list(hour_minute_avg.columns)]
```

```python
# Join hour_avg and hour_minute_avg back to data frame by date (date to index)
energy_dt_3 = pd.merge(energy_dt_2, hour_avg, left_on='hour', right_on='hour')
energy_dt_3 = pd.merge(energy_dt_3, hour_minute_avg, left_on='hour_minute', right_on='hour_minute')

# Save energy_dt with features
with open('datasets/energy_dt_v2.pickle', 'wb') as file:
    pickle.dump(energy_dt_3, file)
```
